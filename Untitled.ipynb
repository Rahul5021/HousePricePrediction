```python
df = df.drop(columns=["PoolQC", "MiscFeature", "Fence", "Alley"])
```

```python
df.info()
```

```python
df[["MasVnrType", "MasVnrArea"]]
```

```python
# Replace NaN values in MasVnrType with "None"
df["MasVnrType"].fillna("None", inplace=True)
```

```python
print(df["MasVnrType"].isnull().sum())
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
print(df["MasVnrArea"].isnull().sum())
```

```python
missing_masvnr = df[df["MasVnrArea"].isnull()][["MasVnrType", "MasVnrArea"]]
print(missing_masvnr)
```

```python
# filling the missing values with 0
df["MasVnrArea"].fillna(0, inplace=True)
```

```python
print(df["LotFrontage"].isnull().sum())
```

```python
# Fill missing 'LotFrontage' values with the median of the column to handle missing data.
df["LotFrontage"].fillna(df["LotFrontage"].median(), inplace=True)
```

```python
df.info()
```

```python
df["BsmtQual"].value_counts()
```

```python
# Handling missing values for categorical columns
df["BsmtQual"].fillna("None", inplace=True)
df["BsmtCond"].fillna("None", inplace=True)
df["BsmtExposure"].fillna("None", inplace=True)
df["BsmtFinType1"].fillna("None", inplace=True)
df["BsmtFinType2"].fillna("None", inplace=True)

# Handling missing values for numeric columns
df["BsmtFinSF1"].fillna(0, inplace=True)
df["BsmtFinSF2"].fillna(0, inplace=True)
df["BsmtUnfSF"].fillna(0, inplace=True)
```

```python
df["Electrical"].value_counts()
```

```python
print(df["Electrical"].isnull().sum())
```

```python
df["Electrical"].fillna(df["Electrical"].mode()[0], inplace = True)
```

```python
df[["Fireplaces","FireplaceQu"]]
```

```python
df["FireplaceQu"].fillna("None", inplace=True)
```

```python
# Fill missing values in 'GarageType', 'GarageFinish', 'GarageQual', and 'GarageCond' with 'No Garage'
df["GarageType"].fillna("No Garage", inplace=True)
df["GarageFinish"].fillna("No Garage", inplace=True)
df["GarageQual"].fillna("No Garage", inplace=True)
df["GarageCond"].fillna("No Garage", inplace=True)

# Fill missing values in 'GarageYrBlt' with 0 (no garage built)
df["GarageYrBlt"].fillna(0, inplace=True)
```
