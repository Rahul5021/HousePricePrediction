```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoder to encode MSZoning
ohe = OneHotEncoder(drop='first', sparse_output=False)
encoded_mszoning = ohe.fit_transform(df[['MSZoning']])

# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_mszoning, columns=ohe.get_feature_names_out(['MSZoning']))

# Concatenate with the original dataframe (drop original 'MSZoning' column)
df = pd.concat([df, encoded_df], axis=1)
df.drop(['MSZoning'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# OrdinalEncoder to encode LotShape
lotshape_order = ['IR3', 'IR2', 'IR1', 'Reg']  # Based on the natural order
encoder = OrdinalEncoder(categories=[lotshape_order])

# Apply encoding
df['LotShape_encoded'] = encoder.fit_transform(df[['LotShape']])
df.drop(['LotShape'], axis=1, inplace=True)
```

```python
# OrdinalEncoder to encode LandContour
landcontour_order = ['Low', 'Lvl', 'Bnk', 'HLS']  # Based on the natural order
encoder = OrdinalEncoder(categories=[landcontour_order])

# Apply encoding
df['LandContour_encoded'] = encoder.fit_transform(df[['LandContour']])
df.drop(['LandContour'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the mapping for the ordinal columns
ordinal_mapping = {
    'ExterQual': ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Quality scale for exterior material
    'BsmtQual': ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Basement quality scale
    'BsmtCond': ['None', 'Po', 'Fa', 'TA', 'Gd']  # Basement condition scale
}

# Initialize OrdinalEncoder
oe = OrdinalEncoder()

# Apply OrdinalEncoder to the specified columns
df['ExterQual'] = oe.fit_transform(df[['ExterQual']])
df['BsmtQual'] = oe.fit_transform(df[['BsmtQual']])
df['BsmtCond'] = oe.fit_transform(df[['BsmtCond']])

# Check the results after ordinal encoding
print(df[['ExterQual', 'BsmtQual', 'BsmtCond']].head())
```

```python
from sklearn.preprocessing import OneHotEncoder

# List of nominal columns you want to apply OneHotEncoding to
nominal_columns = [
    'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 
    'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 
    'MasVnrType', 'Foundation', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'GarageType', 
    'GarageFinish', 'GarageQual', 'GarageCond'
]

# Initialize OneHotEncoder (drop the first column to avoid multicollinearity)
ohe = OneHotEncoder(drop='first', sparse_output=False)

# Apply OneHotEncoder to the nominal columns and convert the result into a DataFrame
encoded_nominal = ohe.fit_transform(df[nominal_columns])

# Create a DataFrame for the encoded nominal columns with proper column names
encoded_df = pd.DataFrame(encoded_nominal, columns=ohe.get_feature_names_out(nominal_columns))

# Concatenate the encoded nominal columns back to the original DataFrame (excluding the original nominal columns)
df_encoded = pd.concat([df.drop(columns=nominal_columns), encoded_df], axis=1)

# Check the results after one-hot encoding
print(df_encoded.head())
```

```python
q1 = df["LotFrontage"].quantile(0.25)
q3 = df["LotFrontage"].quantile(0.75)
IQR = q3 - q1

lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

outliers = df[(df['LotFrontage'] < lower_bound) | (df['LotFrontage'] > upper_bound)]
outliers
```

```python
q1 = df["LotArea"].quantile(0.025)
q3 = df["LotArea"].quantile(0.075)
IQR = q3 - q1

lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

outliers = df[(df['LotArea'] < lower_bound) | (df['LotArea'] > upper_bound)]
outliers
```

```python
# Calculate 97.5th percentile
lot_area_threshold = df['LotArea'].quantile(0.975)

# Cap values at the 97.5th percentile
df['LotArea'] = np.where(df['LotArea'] > lot_area_threshold, lot_area_threshold, df['LotArea'])
```

```python
saleprice_column = df.pop("SalePrice")
df['SalePrice'] = saleprice_column
df.head()
```
